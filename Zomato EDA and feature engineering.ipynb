```python
# combining Country Code from df and Country code from database
final_df = pd.merge(df,dataset,on= 'Country Code',how = 'left')
final_df
```

```python
# To check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
Country_names = final_df.Country.value_counts().index
Country_names
```

```python
Country_val= final_df.Country.value_counts().values
Country_val
```

```python
plt.pie(Country_val[:3],labels = Country_names[:3],autopct = "%0.2f%%")
```

### Observation: 
India seems to be the Country with highest number of Zomato Orders followed by USA and Uk.



```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns= {0:'Rating Count'})
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"] =(25,8)
sns.barplot(x= "Aggregate rating", y= "Rating Count",hue = "Rating color",data = ratings,palette =['blue','red','orange','yellow','green','green'])
```

### Observations 
Here,after observing overall ratings, The average rating seems to be between 2.9 to 3.9.
There are more of 0 ratings compared to all the other ratings.


```python
sns.countplot(x= "Rating color",data= ratings,palette =['blue','red','orange','yellow','green','green'])
```
